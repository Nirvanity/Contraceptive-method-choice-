## Model building

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Contraceptive method used', axis=1)
y = df['Contraceptive method used']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
rfc = RandomForestClassifier(n_estimators=10)
```

```python
rfc.fit(x_train, y_train)
```

```python
predictions = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

### SVM 

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train, y_train)
```

```python
pred = svc.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```

### GridSearchCV

```python
param_grid = {'C': [0.1, 1, 2, 3, 10, 20, 30, 40, 50, 60, 70, 80, 80, 100, 200, 300, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 0.0001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(classification_report(y_test, grid_pred))
```
